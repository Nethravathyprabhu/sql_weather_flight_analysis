```python


# Set the URL
url = "https://meteostat.p.rapidapi.com/point/daily"

# Set lat and lon
#lats = ["40.776863"]
#lons = ["-73.8740"]
lats  = ["40.6413","40.776863","40.689491","39.8730","38.9523","38.8513"]
lons = ["-73.7781","-73.8740","-74.1745","-75.2437","-77.4586","-77.0360"]
faas = ["JFK","LGA","EWR","PHL","IAD","DCA"]
# Set start and end dates
start = "2016-01-01"
end = "2016-01-31"

# Set unit of measurement
unit = "metric"

# Set API key (defined already)
api_key = os.getenv('x-rapidapi-key')
headers = {
    "X-RapidAPI-Host": "meteostat.p.rapidapi.com",
    "X-RapidAPI-Key": api_key
}

# Create empty dataframe, will be used to append each location's weather data
weather_df = pd.DataFrame([])

# Loop through all locations
for lat, lon , faa in zip(lats, lons,faas):
    # Create final URL
    url_f = f"{url}?lat={lat}&lon={lon}&start={start}&end={end}&units={unit}"
    print(url_f)  # Print the URL for debugging
    
    # Request data from URL
    r = requests.get(url_f, headers=headers)
    
    # Optional: Pause to avoid hitting rate limits
    time.sleep(1) 
    
        # Decode response with json decoder
    weather_temp = r.json()

        # Flatten json response
    weather_temp_df = pd.json_normalize(weather_temp,
                                            sep="_",
                                             record_path="data",
                                                #meta=[["meta", "generated"], ["meta", "stations"]],
                                                record_prefix="weather_"
        )
    weather_temp_df['lat'] = lat
    weather_temp_df['lon'] = lon
    weather_temp_df['faa'] = faa
        # Concatenate dataframes
    weather_df = pd.concat([weather_df, weather_temp_df], ignore_index=True)
        
# Print final dataset weather_df
weather_df.head()
```
